```python
import numpy as np
import pandas as pd
```

```python
steuersatz = .25
freibetrag = 12000
```

```python
jahresgehalt = np.arange(1000, 121000, 1000)

steuerlast_p = lambda p : ((jahresgehalt - p*freibetrag) > 0) * (jahresgehalt - p*freibetrag) * steuersatz
steuerlast = steuerlast_p(1)
nettojahresgehalt = jahresgehalt - steuerlast
df = pd.DataFrame({'Brutto-Jahresgehalt' : jahresgehalt,
                   'Steuerlast' : steuerlast,
                   'Netto-Jahresgehalt' : nettojahresgehalt,
                   'Realer Steuersatz' : steuerlast / jahresgehalt * 100})
for p in range(2,6):
    steuerlast = steuerlast_p(p)
    nettojahresgehalt = jahresgehalt - steuerlast
    df['Steuerlast p=%d' %(p)] = steuerlast
    df['Netto-Jahresgehalt p=%d' %(p)] = nettojahresgehalt
    df['Realer Steuersatz p=%d' %(p)] =  steuerlast / jahresgehalt * 100
    
df.head()
```

```python
df.plot(x='Brutto-Jahresgehalt',
        y=['Steuerlast', 'Steuerlast p=2', 'Steuerlast p=3', 'Steuerlast p=4', 'Steuerlast p=5'],
        ylabel='Steuerlast []') ;
```

```python
df.plot(x='Brutto-Jahresgehalt',
        y=['Realer Steuersatz', 'Realer Steuersatz p=2', 'Realer Steuersatz p=3', 'Realer Steuersatz p=4', 'Realer Steuersatz p=5'],
        ylabel='Realer Steuersatz [%]') ;
```
