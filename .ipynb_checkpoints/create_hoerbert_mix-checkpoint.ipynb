```python
import json
import youtube_dl
import matplotlib.pyplot as plt

from pathlib import Path
from slugify import slugify
```

```python
with open('hoerbert.json', 'r', encoding='utf-8') as f:
    hoerbert = json.load(f)

destination = Path('sd_card')
```

```python
for idb, button in enumerate(hoerbert):
    for ids, song in enumerate(hoerbert[button]):
        
        song_name = '%d.mp3' % (ids)
        song_path = destination / str(idb) / song_name
        
        print('\n%s\n%s\n' % (song['Name'], '=' * len(song['Name'])))
        
        ydl_opts = {
            'format': 'bestaudio/best',
            'outtmpl': str(song_path),
            'postprocessors': [{
                'key': 'FFmpegExtractAudio',
                'preferredcodec': 'mp3',
                'preferredquality': '192',
            }]
        }
        with youtube_dl.YoutubeDL(ydl_opts) as ydl:
            ydl.cache.remove()
            ydl.download([song['Link']])
```

```python
colors = ['340B08', 'CC0904', '03307C', 
          '6CBA24', 'FED633', '128B9F',
          'A0CEFB', 'FE884A', '005B25']
coords = [(0, 2), (2.2, 2), (4.4, 2),
          (0, 1), (2.2, 1), (4.4, 1),
          (0, 0), (2.2, 0), (4.4, 0),]
cm = 1/2.54
fig, ax = plt.subplots(figsize=(28*cm,20*cm));

for idb, button in enumerate(hoerbert):
        ax.plot(coords[idb][0],
                coords[idb][1],
                marker='o',
                markersize=31,
                color='#' + colors[idb])
        for ids, song in enumerate(hoerbert[button]):
            short_name = (song['Name'][:30] + '…') if len(song['Name']) > 30 else song['Name']
            song_name = '%d: %s' %(ids+1, short_name)
            ax.text(coords[idb][0]-0.15,
                    coords[idb][1] - ids*0.08 -0.3,
                    song_name,
                    fontsize=10)
        
ax.set_xlim(-0.5, 6.8);
ax.set_ylim(-0.8, 2.5);
ax.axis('off')

fig.savefig(destination / 'index.jpg', dpi=300)
```
